```python
# %% [markdown]
# # Análise Exploratória de Dados - Churn Prediction

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append('..')

from src.data.make_dataset import DataLoader

# Carregar dados
loader = DataLoader()
df = loader.load_data()

# %% [markdown]
# ## Visão Geral dos Dados

# %%
print(f"Shape: {df.shape}")
print(f"\nColunas: {df.columns.tolist()}")
print(f"\nInfo:")
df.info()

# %% [markdown]
# ## Distribuição da Variável Alvo (Churn)

# %%
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

df['Churn'].value_counts().plot(kind='bar', ax=axes[0])
axes[0].set_title('Distribuição de Churn')
axes[0].set_xticklabels(['Não', 'Sim'], rotation=0)

df['Churn'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%',
                                labels=['Não', 'Sim'], colors=['green', 'red'])
axes[1].set_title('Proporção de Churn')

plt.tight_layout()
plt.show()

# %% [markdown]
# ## Análise de Variáveis Numéricas

# %%
numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

for i, col in enumerate(numerical_cols):
    # Histograma
    df[col].hist(ax=axes[0, i], bins=30, edgecolor='black')
    axes[0, i].set_title(f'Distribuição de {col}')

    # Boxplot por Churn
    df.boxplot(column=col, by='Churn', ax=axes[1, i])
    axes[1, i].set_title(f'{col} por Churn')

plt.tight_layout()
plt.show()
```
